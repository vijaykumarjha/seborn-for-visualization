```python
g = sns.FacetGrid(titanic, col = "Survived", size = 6)
g.map(sns.kdeplot,"Age",shade = True)
plt.show()
```

```python
a = sns.FacetGrid(titanic, col = "Pclass", size = 6)
a.map(sns.kdeplot,"Fare",shade = True)
```

```python
b = sns.FacetGrid(titanic, col= "Survived", size = 6)
b.map(sns.kdeplot,"Pclass", shade = True)
```

```python
c = sns.FacetGrid(titanic, col = "Pclass", size = 6)
c.map(sns.kdeplot, "Age", shade = True)
sns.despine(left = True, bottom = True)
```

```python
d = sns.FacetGrid(titanic,col="Age",size =6)
d.map(sns.kdeplot, "Pclass", shade = True)
sns.despine(left = True, bottom = True)
```

```python
e= sns.FacetGrid(titanic,col ="Survived", row = "Pclass", hue = "Sex", size = 6)
e.map(sns.kdeplot,"Age", shade = True)
e.add_legend()
plt.show()
```
